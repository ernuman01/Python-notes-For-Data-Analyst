to change the color categorically

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, hue = "sex", palette = "rocket")

plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, hue = "sex", style = "sex", palette = "rocket", markers = ["o",">"])

plt.show()
```
