```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname = r'arman.jpg'

# opening image using PIL

image = Image.open(fname).convert('L')

#mapping image using PIL
plt.imshow(image, cmap = 'Blues')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname = r'pro.jpg'

# opening image using PIL

image = Image.open(fname).convert('L')

plt.title("Ardaam")  # to use title in pic

plt.xlabel("Year")
plt.ylabel("Height")
#mapping image using PIL
plt.imshow(image, cmap = 'coolwarm')
plt.show()
```

# 4.PIE Chart

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["English", "Hindi","Math", "Science"]
z = ["red", "blue", "black","pink"]

plt.pie(x, labels = y, colors = z)
plt.show()
```
